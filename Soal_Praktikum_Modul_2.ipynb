$$

P(\text{flu} \ | \ \text{vaksin}) &= ?

$$

B. Dari data di atas, berapa peluang seseorang akan mengidap flu apabila **belum** divaksin?

````python
n_flu_dan_vaksin = 200
n_flu_dan_tidak_vaksin = 900

n_tidak_flu_dan_vaksin = 4000
n_tidak_flu_dan_tidak_vaksin = 9000
````

## **Case 3**: Sebuah perusahaan IT memiliki dua divisi, Divisi Pengembangan dan Divisi Pemasaran, yang keduanya mempekerjakan para profesional IT. Dalam Divisi Pengembangan, 30% dari pekerja memiliki pengalaman kerja kurang dari 3 tahun, sedangkan di Divisi Pemasaran, 70% dari pekerja memiliki pengalaman kerja kurang dari 3 tahun.

<a name='3'></a>
### Soal 3.
A. Jika seorang pekerja dipilih secara acak dan ternyata memiliki pengalaman kerja kurang dari 3 tahun, berapa probabilitas bahwa pekerja tersebut berasal dari Divisi Pengembangan? <br>
B. Dengan mempertimbangkan informasi dari Divisi Pengembangan dan Divisi Pemasaran seperti yang diberikan sebelumnya, jika diketahui bahwa sebuah proyek baru memerlukan tim yang terdiri dari pekerja dengan pengalaman kerja kurang dari 3 tahun, dan satu pekerja dipilih secara acak dari keseluruhan pekerja yang memenuhi kriteria ini, berapa rasio antara probabilitas pekerja tersebut berasal dari Divisi Pengembangan terhadap probabilitas pekerja tersebut berasal dari Divisi Pemasaran?

````python
P_exp_dev = 0.3  # Probabilitas pengalaman < 3 tahun jika dari divisi pengembangan
P_exp_marketing = 0.7  # Probabilitas pengalaman < 3 tahun jika dari divisi pemasaran
P_dev = 0.5  # Probabilitas memilih seorang pekerja dari divisi pengembangan
P_marketing= 0.5  # Probabilitas memilih seorang pekerja dari divisi pemasaran
````

### A. Jika seorang pekerja dipilih secara acak dan ternyata memiliki pengalaman kerja kurang dari 3 tahun, berapa probabilitas bahwa pekerja tersebut berasal dari Divisi Pengembangan?

````python
#rumus: P(dev|exp<3) = (P(exp<3|dev) * P(dev)) / P(exp<3)
P_exp_less_than_3 = P_exp_dev * P_dev + P_exp_marketing * P_marketing
P_dev_with_exp_less_than_3 = (P_exp_dev * P_dev) / P_exp_less_than_3

Soal_3_A = P_dev_with_exp_less_than_3

print(f"3A. Jawaban : {Soal_3_A:.2f}")
````

Solver:
$$
\begin{align*}
P(\text{A} \ | \ \text{B}) &= \cfrac{P(\text{B} | \text{A})\times P(\text{A})}{P(\text{B})}
\end{align*}
$$

<br>

$$
\begin{align*}
P(\text{dev} \ | \ \text{exp < 3}) &= \cfrac{P(\text{exp < 3} | \text{dev})\times P(\text{dev})}{P(\text{exp < 3})}
\end{align*}
$$



### B. Rasio probabilitas pekerja berasal dari Divisi Pengembangan terhadap Divisi Pemasaran

````python
ratio_dev_to_marketing = (P_exp_dev * P_dev) / (P_exp_marketing * P_marketing)

Soal_3_B = ratio_dev_to_marketing
print(f"3B. Jawaban : {Soal_3_B:.2f}")
````

Solver:
$$
\begin{align*}
\frac{P(\text{dev} | \text{exp} < 3)}{P(\text{marketing} | \text{exp} < 3)}
\end{align*}
$$

<br>

$$
\begin{align*}
P(\text{marketing} | \text{exp}<3) &= \cfrac{P(\text{exp}<3 | \text{marketing}) \times P(\text{marketing})}{P(\text{exp}<3)}
\end{align*}
$$



**Penjelasan Pengerjaan Soal 3** <br>

untuk mengerjakan no 3, saya menggunakan teori Bayes, karena teori ini memberikan cara untuk menghitung probabilitas suatu kejadian dengan mempertimbangkan informasi atau kejadian lain yang telah diketahui terjadi yang sangat relevan untuk kasus ini, di mana kita ingin mengetahui probabilitas  seorang pekerja berasal dari divisi pengembangan diketahui memiliki pengalaman kerja kurang dari 3 tahun. P_dev = 0.5  # Probabilitas memilih seorang pekerja dari divisi pengembangan

### **Soal 5**: Berapa probabilitas sebuah kemasan perlu diganti, asumsikan bahwa probabilitas sebuah sekrup cacat adalah 0.01 dan setiap kemasan berisi 10 sekrup?

Clue: $P(X>1)$ bisa dicari dengan
$$P(X>1) = 1 - P(X=0) - P(X=1)$$

````python
from scipy.stats import binom

p_cacat = 0.01  # probabilitas sekrup cacat
n_sekrup_per_kemasan = 10  # jumlah sekrup per kemasan

# probabilitas semua 10 sekrup lolos (tidak cacat)
p_semua_lolos = binom.pmf(0, n_sekrup_per_kemasan, p_cacat)

# P=probabilitas setidaknya satu sekrup cacat (kemasan perlu diganti)
p_ganti_kemasan = 1 - p_semua_lolos

Soal_5 = p_ganti_kemasan
print(Soal_5)
````

**Penjelasan Pengerjaan Soal 5** <br>

1. soal ini diselesaikan menggunakan konsep distribusi binomial. Distribusi binomial digunakan untuk menghitung probabilitas mendapatkan sejumlah sukses (dalam konteks ini, sekrup cacat) dalam sejumlah percobaan independen (jumlah sekrup dalam satu kemasan), dengan probabilitas sukses tetap di setiap percobaan. Soal meminta untuk menghitung probabilitas bahwa sebuah kemasan perlu diganti, yang terjadi jika ada setidaknya satu sekrup cacat dalam kemasan. Setiap kemasan berisi 10 sekrup, dan probabilitas sekrup cacat diasumsikan 0.01.

2. dalam konteks soal ini, ada dua kemungkinan yang saling melengkapi:
(1) Semua sekrup dalam satu kemasan lolos uji kualitas (tidak ada yang cacat).
(2) Setidaknya satu sekrup dalam satu kemasan tidak lolos uji kualitas (ada yang cacat).
karena banyak kemungkinan kombinasi sekrup cacat dalam satu kemasan, saya menggunakan pendekatan menghitung probabilitas kejadian nomor 1, lalu mengurangkannya dari 1 untuk mendapatkan probabilitas kejadian nomor 2.

### Rumus Distribusi Binomial <br>
Probabilitas mendapatkan k sukses dalam n percobaan binomial diberikan oleh:
$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

di mana:
- n adalah jumlah percobaan (jumlah sekrup per kemasan),
- k adalah jumlah sukses yang diinginkan (jumlah sekrup cacat),
- p adalah probabilitas sukses per percobaan (probabilitas sekrup cacat),
- $$
\binom{n}{k}
$$
adalah koefisien binomial, yang menghitung jumlah cara
k sukses dapat terjadi dalam n percobaan.

Hitung <br>

````python
# Solver soal 6

# Data skor teknikal dan skor artistik
skor_teknik_x = np.array([96, 88, 92, 90, 94, 89, 93, 87, 95, 91, 90, 92, 89, 96, 88, 92, 89, 87, 96, 88])
skor_artistik_y = np.array([91, 87, 89, 94, 90, 85, 88, 92, 93, 86, 91, 88, 90, 95, 89, 91, 86, 90, 95, 89])

# A. Expectation (Nilai Harapan)
expectation_teknik = np.mean(skor_teknik_x)
expectation_artistik = np.mean(skor_artistik_y)

print("6A. Expectation:")
print("   Skor Teknik:", expectation_teknik)
print("   Skor Artistik:", expectation_artistik)

# B. Variance (Varians)
variance_teknik = np.var(skor_teknik_x, ddof=1)  # menggunakan ddof=1 karena ini sampel, bukan populasi
variance_artistik = np.var(skor_artistik_y, ddof=1)  # menggunakan ddof=1 karena ini sampel, bukan populasi

print("\n6B. Variance:")
print("   Skor Teknik:", variance_teknik)
print("   Skor Artistik:", variance_artistik)

# C. Covariance Matrix (Matriks Kovarians)
covariance_matrix = np.cov(skor_teknik_x, skor_artistik_y, ddof=1)  # menggunakan ddof=1 karena ini sampel, bukan populasi
print("\n6C. Covariance Matrix:")
print(covariance_matrix)
````
